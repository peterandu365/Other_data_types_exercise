```python

# Question 2:

# 1. 10 countries with the lowest infant mortality rates
# 2. 10 cities with the largest population
# 3. name and country of a) longest river, b) largest lake and c) airport at highest elevation

my_result = {}

for country in tree.iterfind('country'):
    for city in country.iterfind('city'):
        recent_population = {}

        try:
            recent_population['population'] = int(city.findall("./population")[-1].text) # find all, more than one
        except:
            pass
            # recent_population['population'] = None
        # if recent_population is not None:
        my_result[city.findtext("name")] = recent_population

df_city_population = pd.DataFrame(my_result).T
df_city_population.dropna().sort_values(by = "population", ascending = False).head(10)
```

```python
#Q3 - 1:

# 1. 10 countries with the lowest infant mortality rates
# 2. 10 cities with the largest population
# 3. name and country of a) longest river, b) largest lake and c) airport at highest elevation

my_result = {}


for river in tree.iterfind('river'):
    river_dict={}

    try: 
        river_dict["length"] = river.findtext("length")
        river_dict["country"] = river.attrib["country"]
    except:
        pass

    my_result[river.findtext("name")] = river_dict

df = pd.DataFrame(my_result).T

df.dropna().sort_values(by = "length", ascending = False).head(1)

```

```python
# Q - 3:


my_result = {}


for airport in tree.iterfind('airport'):
    airport_dict={}

    try: 
        airport_dict["elevation"] = airport.findtext("elevation")
        airport_dict["country"] = airport.attrib["country"]
    except:
        pass

    my_result[airport.findtext("name")] = airport_dict

df = pd.DataFrame(my_result).T

df.dropna().sort_values(by = "elevation", ascending = False).head(1)
```
